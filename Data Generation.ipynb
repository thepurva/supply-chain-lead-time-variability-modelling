```python
import pandas as pd
import numpy as np
import datetime

# 1. how much data we have? (we take 5000)
num_orders = 5000
np.random.seed(42) 

# 2. generate basic features.
order_ids = [f"ORD_{i:04d}" for i in range(1, num_orders + 1)]

# 500 days data towards from 1 january 2024 
start_date = datetime.date(2024, 1, 1)
date_list = [start_date + datetime.timedelta(days=int(np.random.randint(0, 500))) for _ in range(num_orders)]

suppliers = np.random.choice(['Supplier_A', 'Supplier_B', 'Supplier_C'], size=num_orders, p=[0.5, 0.3, 0.2])
transport_modes = np.random.choice(['Sea', 'Air', 'Road'], size=num_orders, p=[0.4, 0.2, 0.4])
weather_conditions = np.random.choice(['Clear', 'Rainy', 'Stormy'], size=num_orders, p=[0.7, 0.2, 0.1])
distance_km = np.random.randint(100, 5000, size=num_orders)

# 3 base lead time
scheduled_lead_time = np.round(distance_km / 400) + 2 # समजा रोज ४०० किमी गाडी चालते + २ दिवस पॅकिंगचे

# 4.  (Delays) & (Logic)
actual_lead_time = scheduled_lead_time.copy()

for i in range(num_orders):
    delay = 0
    #  transport by ocean then delay will be 6 to 12 days.
    if transport_modes[i] == 'Sea' and weather_conditions[i] == 'Stormy':
        delay += np.random.randint(6, 12)
    # if supplier is  'Supplier_C' then delay will be 2 to 5 days.
    if suppliers[i] == 'Supplier_C':
        delay += np.random.randint(2, 5)
    # if weather are not good then road or traffic delay 1 to 3 days.
    if transport_modes[i] == 'Road' and weather_conditions[i] == 'Rainy':
        delay += np.random.randint(1, 4)
        
    # RAndom difference (Noise)
    delay += np.random.randint(-1, 3)
    
    actual_lead_time[i] += delay

# 5. generate data frame
df = pd.DataFrame({
    'Order_ID': order_ids,
    'Order_Date': date_list,
    'Supplier': suppliers,
    'Transport_Mode': transport_modes,
    'Weather': weather_conditions,
    'Distance_KM': distance_km,
    'Scheduled_Lead_Time': scheduled_lead_time,
    'Actual_Lead_Time': actual_lead_time
})

# 6. repair negative days (atleast 1 day)
df['Actual_Lead_Time'] = df['Actual_Lead_Time'].clip(lower=1)

# 7.Export the data.
df.to_csv('supply_chain_data.csv', index=False)
print("Supply chain dataset successfully generated !")
```
